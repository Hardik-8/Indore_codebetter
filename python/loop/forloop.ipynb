```python
rows=5
i=1
while i<=rows:
    space=rows-i
    while space>0:
        print("  ",end="")
        space-= 1

    star=1
    while star<=i:
        print("*   ",end="")
        star+=1
    print()
    i += 1
```

```python
rows=5
i=1
while i<=rows:
    space=rows-i
    while space>0:
        print("  ", end="")
        space-=1

    star=1
    while star<=(2*i-1):
        print("* ", end="")
        star+=1
    print()
    i += 1
```

```python
#wap to print the sum of n!
n=int(input("n:"))
print("n:",n)
sum=0
for i in range(1,n+1):
    j=1
    fact=1
    while j<=2*i:
        fact=fact*j
        j+=1
    sum=sum+fact
print(sum)
```

```python
#wap to print the sum of n!(odd) 1!+3!...n!
n=int(input("n:"))
print("n:",n)
sum=0
for i in range(1,n+1):
    j=1
    fact=1
    while j<=2*i-1:
        fact=fact*j
        j+=1
    sum=sum+fact
print(sum)
```

```python
#wap to print the prime numbers from n
n=int(input("n:"))
print("n:",n)
for i in range (1,n+1):
    j=2
    while j<i:
        if i%j==0:
            break
        j+=1
    if j==i:
        print(i,end=" ")
```
