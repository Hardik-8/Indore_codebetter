20. 1!/2! - 3!/4! + 5!/6! - 7!/8!

```python
sum = 0
sign = 1

for i in range(1, 8, 2):   
    f1 = 1
    for j in range(1, i+1):
        f1 = f1 * j        # i!

    f2 = 1
    for k in range(1, i+2+1):
        f2 = f2 * k       

    sum = sum + sign * (f1 / f2)
    sign = -sign          

print("Sum =", sum)
```
