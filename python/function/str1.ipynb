```python
import random

snakes = {99:3,95:56,85:64,79:60,74:11,67:36,43:19,34:27}
ladders = {4:24,26:98,34:72,40:63,66:94}

player = 0

print("Welcome to Snake and Ladder Game")

while True:
    input("\nPress Enter to roll the dice...")
    dice = random.randint(1, 6)
    print("Dice:", dice)

    if player + dice <= 100:
        player += dice
    else:
        print("Need exact number to reach 100")

    print("You moved to:", player)

    if player in snakes:
        print("Snake bite!")
        player = snakes[player]
        print("Down to:", player)

    elif player in ladders:
        print("Ladder climb!")
        player = ladders[player]
        print("Up to:", player)

    if player == 100:
        print("\n Congratulations! You won the game ")
        break
```
