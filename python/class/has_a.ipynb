```python
class Car:
    def __init__(self):
        #only model color and prize are given
        self.model = ""
        self.color = ""
        self.price = 0
    def display(self):
        print("car - ", self.model, "color - ", self.color, "price - ", self.price)
    def __str__(self):
        return self.model+","+self.color+","+str(self.price)
    
class Person:
    def __init__(self):
        self.name = ""
        self.city = ""
        self.mycar=None
    def display(self):
        print("person - ", self.name, "city - ", self.city ,",",self.mycar)

    def __str__(self):
        return self.name+","+self.city+","+str(self.mycar)
    
p1=Person()
p1.name="John"
p1.city="New York"
p1.mycar="Supra"
p1.display()
# print(p1)

c1 = Car()
c1.model="Toyota"
c1.color="Red"
c1.price=10000
c1.display()

p1.mycar=c1
p1.display()


p2=Person()
p1.name="hardik"
p1.city="indore"

print(c1)
p1.mycar=Car()
p1.mycar.model="Toyota"
p1.mycar.color="Red"
p1.mycar.price=10000
p1.display()
print(p1)

print("-------------------------")
p2=Person()
p2.name="Jane"
p2.city="Los Angeles"
p2.mycar=Car()
p2.mycar.model="Ford"
p2.mycar.color="Blue"
p2.mycar.price=20000
p2.display()


p3=Person()
p3.name="Mike"
p3.city="Chicago"
p3.mycar=c1
p3.display()
```
