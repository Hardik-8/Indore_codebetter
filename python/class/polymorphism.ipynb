```python
class employee:
    def __init__(self):
        self.empid=0
        self.name=""
        self.salary=0
    def show(self):
        print(self.empid,self.name,self.salary)
    def findsalary(self):
        print("salary",self.salary)

class fulltime(employee):
    def __init__(self):
        super().__init__()
        self.basic=0
        self.pf=0
        self.da=0
    def show(self):
        super().show()
        print(self.empid,self.name,self.salary,self.basic,self.pf,self.da)

    # def findsalary(self):
    #     return super().findsalary()
    def findsalary(self):
        self.salary=self.basic+self.pf-self.da
        print(self.salary)


ft1=fulltime()
ft1.name="hardik"
ft1.empid=1
ft1.salary=0
ft1.show()
ft1.basic=67801
ft1.pf=8976
ft1.da=7621
ft1.findsalary()
ft1.show()

```

```python
class part_time(employee):
    def __init__(self):
        super().__init__()
        self.hour=0
        self.rate=0

    def show(self):
        super().show()
        print("PT",self.hour,self.rate)

    def findsalary(self):
        self.salary=self.hour*self.rate
        print("salary",self.salary)

ft2=part_time()
ft2.name="hardik"
ft2.empid=2
ft2.salary=0
ft2.show()
ft2.hour=8
ft2.rate=100
ft2.findsalary()
ft2.show()
```
