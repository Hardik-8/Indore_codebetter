```python
class circle:
    def __init__(self):
        self.radius = 0
    def diameter(self):
        d=2*self.radius 
        print(f"the diameter of the circle is {d}cm")
    def circumference(self):
        c=2*3.14*self.radius
        print(f"the circumference of the circle is {c}cm")
    def area(self):
        a=3.14*(self.radius**2)
        print(f"the area of the circle is {a}")

c1=circle()
c1.radius=2
c1.diameter()
c1.circumference()
c1.area()
```
