```python
rec = []
num = int(input("Enter how many student data do you want to save: "))

for i in range(1, num+1):
    roll_no = int(input("Enter Roll Number : "))
    name = input("Enter Name : ")
    marks1 = int(input("Enter physics marks : "))
    marks2 = int(input("Enter chemistry marks : "))
    marks3 = int(input("Enter maths marks : "))
    total = marks1 + marks2 + marks3
    per = total / 3
    ls = [roll_no, name, marks1, marks2, marks3, total, per]
    rec.append(ls)

while True:
    option = int(input("""Press 1 for check all records
Press 2 for check who are passed in first division
Press 3 for check who are passed in second division
Press 4 for check who are passed in third division
Press 5 for check who is failed in one subject
Press 6 for check who is failed in two subject
Press 7 for check who is failed in all subject
Press 8 for search record by Roll No
Press 9 for exit
Enter your choice: """))

    print("\nRoll\tName\tPhy\tChem\tMath\tTotal\tPer")

    if option == 1:
        for x in rec:
            print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 2:
        for x in rec:
            if x[6] >= 80:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 3:
        for x in rec:
            if 60 <= x[6] < 80:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 4:
        for x in rec:
            if 33 <= x[6] < 60:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 5:
        for x in rec:
            fail = 0
            if x[2] < 33: fail += 1
            if x[3] < 33: fail += 1
            if x[4] < 33: fail += 1
            if fail == 1:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 6:
        for x in rec:
            fail = 0
            if x[2] < 33: fail += 1
            if x[3] < 33: fail += 1
            if x[4] < 33: fail += 1
            if fail == 2:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 7:
        for x in rec:
            if x[2] < 33 and x[3] < 33 and x[4] < 33:
                print(x[0],"\t",x[1],"\t",x[2],"\t",x[3],"\t",x[4],"\t",x[5],"\t",round(x[6],2))

    elif option == 8:
        r = int(input("Enter Roll Number to Search: "))
        for i in rec:
            z=i[0]
            if z==r:
                print("The name of student is ",i[1],"Marks in physics are ",i[2],"Marks in Chemistry are ",i[3],"Marks in maths are ",i[4],"Total percentage are ",i[6])

    elif option == 9:
        print("Thankyou for using this program")
        break

    else:
        print("Invalid option, Try again")
```
