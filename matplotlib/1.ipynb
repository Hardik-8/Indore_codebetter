```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=["bca",'bba','mba','mca','btech']
y=[140000,120000,150000,130000,160000]
# plt.bar(x,y)
plt.plot(x,y,color='blue',marker='o',label='fees')
plt.xlabel("Course")
plt.ylabel("Fees")
plt.title("Structure")
plt.legend()
plt.grid()
plt.show()
```

```python
course = ['bca','bba','bcom', 'ba', 'bsc']
fee = [15000, 10000, 16000 , 12000 , 14000]
student = [1000, 2000, 7000, 4000, 5000]
plt.plot(course, fee, color= 'violet', marker = 'o', label= "course by fees",linewidth=1,ls='--',mfc='brown',ms=8)   # by default will give line plot
plt.plot(course, student, color= 'blue', marker= 'o', label= "course by student",ls=':')
plt.xlabel("Courses",fontdict={'fontfamily':'arial','color':'yellow','fontsize':20})
plt.ylabel("Fees",fontdict={'fontfamily':'arial','color':'yellow','fontsize':20})
plt.title("bar chart")
plt.grid(True,color='black',ls='--')
plt.legend()
plt.show()
```
