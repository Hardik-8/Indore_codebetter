```python
df=pd.DataFrame({
    'Category':pd.Categorical(['low','medium','high','medium','low']),
    'Value':[100,200,300,400,500]})
df
```

```python
new=df.sort_values(by='Category')
new
```

```python
df_high=df[df['Category']=='high']
df_high
```

```python
df_medium=df[df['Category']=='medium']
df_medium
```

```python
df_low=df[df['Category']=='low']
df_low
```

```python
df['Category']=df['Category'].replace({'low':'very low'})
df
```

```python
category_count=df['Category'].value_counts()
category_count
```

```python
df_encode=pd.get_dummies(df,columns=['Category'])
df_encode
```
