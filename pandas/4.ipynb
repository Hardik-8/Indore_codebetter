```python
import pandas as pd 
```

```python
df=pd.read_csv('employee.csv')
df.head(0)
```

```python
new_df=df[['emp_id','age','salary','name']]
new_df
```

```python
new_df.head(2)
```

```python
new_df.head(2)
```

```python
new_df['ta']=new_df['salary']*0.16
```

```python
new_df['da']=new_df['salary']*0.20
```

```python
new_df
```

```python
new_df['department']=['HR','IT','SALES','FINANCE','ACCOUNTS','IT','IT','SALES','FINANCE','IT']
```

```python
new_df['gross_salary']=new_df['salary']+new_df['ta']+new_df['da']
```

```python
new_df.head(0)
```

```python
# new_df.to_csv("employee2.csv",index=False)
```

```python
new_df['tax']=new_df['gross_salary']*0.05
```

```python
new_df.head(11)
```

```python
new_df['actual_salary']=new_df['gross_salary']-new_df['tax']
```

```python
new_df.head(11)
```

```python
# new_sum=new_df[['ta','da','gross_salary','tax','actual_salary']]
```

```python
tot_sal=new_df['salary'].sum()
```

```python
tot_ta=new_df['ta'].sum()
tot_da=new_df['da'].sum()
tot_gross=new_df['gross_salary'].sum()
tot_tax=new_df['tax'].sum()
tot_actual=new_df['actual_salary'].sum()
```

```python
print("total salary : ",tot_sal)
print("total ta :",tot_ta)
print("total da : ",tot_da)
print("total gross salary : ",tot_gross)
print("total tax : ",tot_tax)
print("total actual salary : ",tot_actual)

```
