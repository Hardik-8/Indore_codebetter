```python
import pandas as pd
```

```python
students={
    "student_id":[1,2,3,4,5],
    "std_name":["John","Jane","Jim","Jill","Jack"],
    "std_age":[20,21,22,23,24],
    "city":["New York","Los Angeles","Chicago","Houston","Miami"],
    "course":["Math","Science","Science","English","Math"],
    "fee":[1000,2000,3000,4000,5000]
}

df=pd.DataFrame(students)
print(df)
```

```python
print(df['fee'].mean())
```

```python
print(df['fee'].sum())
```

```python
print(df['fee'].max())
```

```python
print(df['fee'].min())
```

```python
print(df['fee'].std())
```

```python
print(df['fee'].var())
```

```python
print(df['fee'].median())
```

```python
df.groupby('course')['fee'].sum()
```

```python
df.groupby('course')['fee'].min()
```

```python
df.groupby('course')['fee'].mean()
```

```python
df.groupby('course')['fee'].max()
```

```python
df.groupby('city')['fee'].sum()
```

```python
df.groupby('city')['fee'].max()
```

```python
df.groupby('city')['fee'].var()
```

```python
df.groupby('city')['fee'].mean()
```

```python
df.groupby('city')['fee'].std()
```

```python
df.groupby('city')['fee'].agg({'sum','max','min','mean','count'})
```
