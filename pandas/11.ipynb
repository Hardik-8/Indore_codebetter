```python
import pandas as pd
ind=['a','b','c','d','e']
students={
    "student_id":[101,102,103,104,105],
    "std_name":["John","Jane","Jim","Jill","Jack"],
    "std_age":[20,21,22,23,24],
    "city":["New York","Los Angeles","Chicago","Houston","Miami"],
    "c_id_ ":[101,102,103,104,105]

}

df=pd.DataFrame(students,index=ind)
print(df)

df2=df.reindex(['a','b','c','d'])
print(df2)

df3=df.reindex(columns=['student_id','std_name','std_age','city','c_id_ '])
print(df3)

df4=df.reindex(columns=['student_id','std_name','std_age','city','c_id_ '],fill_value='indore')
print(df4)
```
