```python
import pandas as pd
data = {'Name' : ['Ajay','Vijay','Jay'],
        'Age' : [20,25,50],
        'Salary' : [500000,35467,23453]
}
df=pd.DataFrame(data)
```

```python
def age_group(age):
    if age < 18:
        return 'minor'
    elif age >= 18:
        return 'adult'
    else:
        return 'senior'
```

```python
df['age_group']=df['Age'].apply(age_group)
```

```python
df['salary with bonus']=df['Salary'].apply(lambda x: x*1.10)
df
```

```python
df['tax of salary']=df['Salary'].apply(lambda x: x*0.15)
df
```

```python
df2=df.applymap(lambda x: x*2)
df2
```

```python
df3=pd.DataFrame({
    'Name':['Ajay Mishra','Vijay','Jay'],
    'Age':[20,25,50],
    'Salary':[500000,35467,23453]
})
df3['Initials']=df3['Name'].map(lambda x: ''.join([i[2] for i in x.split()]))
df3
```

```python
df4=pd.DataFrame({
    'A':[1,2,3],
    'B':[40,50,60]
})
```

```python
df4_standardised=df4.transform(lambda x : (x-x.mean())/x.std())
df4_standardised
```
