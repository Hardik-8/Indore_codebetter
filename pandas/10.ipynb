```python
import pandas as pd
```

```python
course={
    "course_id":[401,402,403,404,408],
    "course_name":['BCA','BTECH','MCA','BCOM','MBA'],
    "course_fees":[20000,22000,21290,23435,28900],
    'course_duration':['3 years','4 years','2 years','3 years','2 years']
}

student={
    "std_id":[1,2,3,4,5,6,7,8],
    "name":['raj','ravi','riya','hardik','tanvi','renu','deepak','nihal'],
    "age":[20,22,21,24,22,27,28,25],
    "city":['indore','satna','katni','indore','maihar','bpl','jbp','satna'] ,
    "c_id":[401,402,405,401,402,403,404,407]
}
c_df=pd.DataFrame(course)
s_df=pd.DataFrame(student)

```

```python
c_df
```

```python
s_df
```

```python
# inner
student_course=pd.merge(s_df,c_df,how='inner',left_on='c_id',right_on='course_id')
```

```python
student_course
```

```python
# inner join me agar left table aur right table me koi bhi column ka data same hota hai jaise ki dono me course id and c_id 
# ho to waha pe ham inner join use krte hai 
```

```python
# left join me sab kuch join hota hai jaise ki left table me jo jo bhi data hai vo sara righ vali table me dekhta hai aur agar value nahi hai to 
# null ho jaega 
```

```python
# left
student_course=pd.merge(s_df,c_df,how='left',left_on='c_id',right_on='course_id')
student_course
```

```python
# cross join me multiply hote hai 
```

```python
# right
student_course=pd.merge(s_df,c_df,how='right',left_on='c_id',right_on='course_id')
student_course
```

```python
# outer
student_course=pd.merge(s_df,c_df,how='outer',left_on='c_id',right_on='course_id')
student_course
```

```python
# cross
student_course=pd.merge(s_df,c_df,how='cross')
student_course
```
