```python
import numpy as np 
```

```python
ar1=np.array([10,20,30,40,50])
print(ar1.ndim)
print(ar1.shape)
print(ar1.size)
print(ar1.itemsize)
```

```python
#aggregate functions 
# max()
# min()
# mean()
# median()
# sum()
# var()
# std()
```

```python
print("max",np.max(ar1))
```

```python
print("mean",np.mean(ar1))
```

```python
print("median",np.median(ar1))
```

```python
print("sum",np.sum(ar1))
```

```python
print("sum",np.shape(ar1))
```

```python
print("standard division",np.std(ar1))
```

```python
print("var",np.var(ar1))
```

```python
#variance manually 
avg=np.mean(ar1)
ar2=ar1-avg
ar3=ar2**2
ar4=ar3.sum()
var_res=ar4/ar1.size
std_res=np.sqrt(var_res)
print("variance",var_res)
print("Standard division",std_res)
```
